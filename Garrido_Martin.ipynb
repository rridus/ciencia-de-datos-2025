```python
import matplotlib.pyplot as plt
import numpy as np
from Estadistica import promedio, mediana, desviacion_estandar, rango_intercuartilico

# 1. Obtener datos de velocidad radial (RV)
rv = df_virgo['RV'].values

# 2. Regla de Freeman & Diaconis para los bines [cite: 41]
# Ancho de bin = 2 * IQR * n^(-1/3)
n = len(rv)
iqr_val = rango_intercuartilico(rv.tolist())
h_fd = 2 * iqr_val * (n**(-1/3))

# Usar la mitad del ancho prescrito [cite: 43]
ancho_bin = h_fd / 2

# Crear la lista de bines desde el mínimo al máximo
bines = np.arange(min(rv), max(rv) + ancho_bin, ancho_bin)

# 3. Cálculos estadísticos usando tus funciones
prom_rv = promedio(rv.tolist())
med_rv = mediana(rv.tolist())
std_rv = desviacion_estandar(rv.tolist())

# 4. Gráfico orientado a objetos [cite: 16]
fig, ax = plt.subplots(figsize=(10, 7))
ax.hist(rv, bins=bines, color='lightgray', edgecolor='black', alpha=0.7)

# Línea de Promedio y STD [cite: 44, 48]
ax.axvline(prom_rv, color='black', linestyle='solid', linewidth=2, label='Promedio')
ax.text(prom_rv + 20, ax.get_ylim()[1]*0.9, f'Promedio {int(prom_rv)} km/s', color='black')
ax.text(prom_rv + 20, ax.get_ylim()[1]*0.85, f'(STD={int(std_rv)} km/s)', color='black')
ax.axvline(prom_rv - std_rv, color='green', linestyle='dotted')
ax.axvline(prom_rv + std_rv, color='green', linestyle='dotted')

# Línea de Mediana e IQR [cite: 46, 49, 50]
ax.axvline(med_rv, color='red', linestyle='dashed', linewidth=2, label='Mediana')
ax.text(med_rv - 350, ax.get_ylim()[1]*0.7, f'Mediana = {int(med_rv)} km/s', color='red')
ax.text(med_rv - 350, ax.get_ylim()[1]*0.65, f'IQR = {int(iqr_val)}', color='red')

# Líneas de Cuartiles (Q1 y Q3) estimadas [cite: 50]
# Nota: Puedes calcular Q1 y Q3 con una lógica similar a tu función IQR
lista_ordenada = sorted(rv.tolist())
q1 = lista_ordenada[n // 4]
q3 = lista_ordenada[(n * 3) // 4]
ax.axvline(q1, color='blue', linestyle='dotted')
ax.axvline(q3, color='blue', linestyle='dotted')

ax.set_xlabel('Velocidad Radial (km/s)')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma de Velocidades Radiales - Cúmulo de Virgo')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys

# 1. Ajustar el path e importar con el nombre REAL del archivo
sys.path.append("/content/ciencia-de-datos-2025")
from estadistica_py import mediana, MAD # Usamos el nombre que apareció en tu lista

# 2. Cargar datos del bulbo galáctico
columnas_bulbo = ['ID', 'glon', 'glat', 'vrad', 'fe_h', 'mg_fe', 'o_fe']
df_bulbo = pd.read_csv('ciencia-de-datos-2025/bulbo_galactico.dat', sep='\s+', names=columnas_bulbo, header=None) [cite: 58, 59, 60, 61, 62, 63, 64, 65]

# 3. Crear el gráfico (Estilo Orientado a Objetos)
fig, ax = plt.subplots(figsize=(10, 8)) [cite: 16]

# Gráfico de dispersión de puntos grises pequeños
ax.scatter(df_bulbo['fe_h'], df_bulbo['mg_fe'], s=1, color='gray', alpha=0.5, label='Datos') [cite: 67, 68]
ax.set_xlim(-2.0, 0.75) [cite: 69]
ax.set_ylim(-0.2, 0.6) [cite: 69]

# 4. Dividir en 18 intervalos de metalicidad [Fe/H]
bins_fe = np.linspace(-2.0, 0.75, 19) # 18 intervalos necesitan 19 cortes
puntos_medios = []
curva_mediana = []
curva_p5 = []
curva_p95 = []
curva_mad_pos = []
curva_mad_neg = []

for i in range(len(bins_fe)-1):
    # Filtrar estrellas en el intervalo actual
    mask = (df_bulbo['fe_h'] >= bins_fe[i]) & (df_bulbo['fe_h'] < bins_fe[i+1])
    datos_intervalo = df_bulbo.loc[mask, 'mg_fe'].tolist()

    # Solo si hay más de 5 estrellas [cite: 70]
    if len(datos_intervalo) > 5:
        p_medio = (bins_fe[i] + bins_fe[i+1]) / 2
        puntos_medios.append(p_medio)

        # Cálculos con tus funciones
        m = mediana(datos_intervalo)
        valor_mad = MAD(datos_intervalo)

        curva_mediana.append(m)
        curva_mad_pos.append(m + valor_mad) [cite: 76]
        curva_mad_neg.append(m - valor_mad) [cite: 76]

        # Percentiles (usando lógica de tu función mediana/cuartiles)
        datos_ord = sorted(datos_intervalo)
        curva_p5.append(datos_ord[int(len(datos_ord)*0.05)]) [cite: 73]
        curva_p95.append(datos_ord[int(len(datos_ord)*0.95)]) [cite: 73]

# 5. Graficar las curvas
ax.plot(puntos_medios, curva_mediana, color='red', marker='x', label='Mediana') [cite: 71, 72]
ax.plot(puntos_medios, curva_p5, color='green', linestyle='--', label='Percentil 5%/95%') [cite: 74]
ax.plot(puntos_medios, curva_p95, color='green', linestyle='--') [cite: 74]
ax.plot(puntos_medios, curva_mad_pos, color='magenta', label='Mediana ± MAD') [cite: 76]
ax.plot(puntos_medios, curva_mad_neg, color='magenta') [cite: 76]

ax.set_xlabel('[Fe/H]') [cite: 67]
ax.set_ylabel('[Mg/Fe]') [cite: 67]
ax.legend()
plt.show()
```
